```python
import pandas as pd

data = pd.read_csv('../players_21_clean.csv')
pd.options.display.max_columns = None
```

```python
copy_data = data
```

```python
copy_data
```

```python
copy_data = copy_data.drop(range(9565, 18944))
```

```python
copy_data
```

```python
# create attributes

attributes = copy_data.iloc[:, 33:]
attributes = attributes.drop(['player_traits'], axis=1)

attributes
```

```python
# Load necessary libraries for model

import numpy as np
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.decomposition import PCA
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore')

scaled = StandardScaler()
X = scaled.fit_transform(attributes)

recommendations = NearestNeighbors(n_neighbors=6, algorithm='ball_tree').fit(X)
player_indices = recommendations.kneighbors(X)[1]
```

```python
def get_index(x):
    return copy_data[copy_data['short_name']==x].index.tolist()[0]

def recommend_me(player):
    print('Here are 5 players similar to', player, ':' '\n')
    index = get_index(player)
    for i in player_indices[index][1:]:
            print(copy_data.iloc[i]['short_name'], '\n')

recommend_me("K. De Bruyne")
```

```python
# Save the model

import joblib

joblib.dump(X, '../half-knnplayer.pkl')
```
